```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_excel("/content/drive/MyDrive/student_data.xlsx")
df
```

```python
plt.scatter(df.hours_studied,df.passed,color="red")
plt.xlabel("hours studied",color="red")
plt.ylabel("passed",color="red")
plt.title("hours studied vs passed",color="green")
plt.show()
plt
```

```python
logistic_model =LogisticRegression()
logistic_model.fit(df[["hours_studied"]],df.passed)
```

```python
logistic_model.predict(pd.DataFrame({'hours_studied': [6.99]}))
```

```python
logistic_model.score(df[['hours_studied']], df.passed)
```
